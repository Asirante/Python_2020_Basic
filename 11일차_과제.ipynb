```
import numpy as np
from PIL import Image
from wordcloud import WordCloud
```

```
img = '/content/drive/My Drive/Colab Notebooks/Data/korean.png'
text = open('/content/drive/My Drive/Colab Notebooks/Data/한글.txt').read()
text
```

```
Image.open(img)
```

```
Ip_mask = np.array(Image.open(img))
Ip_mask
```

```
Ip_wc=WordCloud(background_color = 'white',font_path='/content/drive/My Drive/Colab Notebooks/Data/ChosunCentennial_ttf.ttf',mask=Ip_mask,max_font_size=50,contour_width=5,contour_color='green').generate(text)
```

```
Ip_wc.words_
```

```
Ip_wc.to_image()
```
