**내 생일의 최고 기온 데이터 시각화**

```
import matplotlib.pyplot as plt
import csv
```

```
f = open('/content/drive/My Drive/Colab Notebooks/Data/ta.csv')
data = csv.reader(f)
next(data)
```

```
mx_temp = []
for row in data :
  if row[-1] != '' and row[0].split('-')[1] == '09' and row[0].split('-')[2] == '13':
    mx_temp.append(float(row[-1]))
plt.plot(mx_temp)
plt.show()
```

**우리동네 인구구조 시각화**

```
f = open('/content/drive/My Drive/Colab Notebooks/Data/인구데이터.csv')
data = csv.reader(f)
mytown = []
for row in data :
  if '만석동' in row[0] :
    for i in range(3, 104) :
      mytown.append(int(row[i]))
plt.figure(dpi=300)
plt.title('ManSuckDong')
plt.plot(mytown)
plt.show
```
