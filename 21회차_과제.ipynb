```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [1,3,7,5]
plt.plot(y)
```

```python
x = [1,2,6,7]
y = [1,3,7,5]
plt.plot(x,y,'r',marker='o')
```

```python
a = list(range(1,10,2))
b = list(range(-10,10,4))
plt.plot(a,b,'--',marker='o',color = 'g')
```

```python
a = np.arange(-10,10,0.5)
b =  a * a * a
b2 = b * a + a
plt.plot(a,b2,'-',marker='o',color = 'c')
plt.plot(a,b,'--',marker='^',color = 'y')
```

```python
plt.subplot(1,2,1)
plt.plot(a,b2,'-',marker='o',color = 'c')
plt.subplot(1,2,2)
plt.plot(a,b,'--',marker='^',color = 'y')
```

```python
plt.plot(a,b2,'-',marker='o',color = 'c')
plt.figure()
plt.plot(a,b,'--',marker='^',color = 'y')
```

```python
t = np.arange(0,12,0.5)
y1 = np.sin(t)
y2 = np.cos(t)
plt.grid()
plt.plot(t,y1,'c--.',marker='^')
plt.plot(t,y2,'m--.',marker='o')
plt.xlabel('Time')
plt.ylabel('Day')
plt.title('Yes')
plt.legend('12')
```

```python
plt.title('God')
x = np.arange(0,5,1)
y1 = x+1
y2 = x+2
y3 = x+3
y4 = x+4
plt.subplot(1,2,1)
plt.plot(x,y1,'y--',marker='o')
plt.plot(x,y2,'r--',marker='^')
plt.xlabel('x')
plt.ylabel('y')
plt.title('t')
plt.legend('12')
plt.subplot(1,2,2)
plt.plot(x,y3,'m--',marker='s')
plt.plot(x,y4,'c--',marker='d')
plt.xlabel('x')
plt.ylabel('y')
plt.title('t')
plt.legend('34')
```
